<strong><h3>Datos nulos</h3></strong>
<p>Los datos nulos pueden llegar a ser un gran problema en las bases de datos, ya que es información que hace falta y que no es muy conveniente para tomar decisiones de una forma más asegurada. Las formas de intentar obtener un modelo que sea muy fiel es deshacerse de los datos que sean incompletos en caso de ser pocos. El verdadero problema es cuando ya son demasiados porque perdemos casi todo.
<p>En este caso estamos sumando todos los datos que sean nulos para demostrar lo fuerte que puede ser el problema.
   
<p>Para obtener el número de datos nulos utilizamos la función <strong>.isna().sum()</strong>.

```python
lluvias.isna().sum()
```

<strong><h3>Datos por todos lados</h3></strong>
<p>Existen bases que tienen los datos algo desordenados, ya sea por como fueron escritos o por la propia estructura. Normalmente lo que buscamos es tener una vista clara de lo que hay. Para estos casos es cuando utilizamos funciones para agrupar datos, como por ejemplo que aquí se agrupó cada ciudad para tener un promedio de datos de forma simplificada con la función <strong>.groupby("columna").agg(parámetros.sum())</strong>

```python
# Obtener los datos por ciudad
lluvias.groupby("Location").agg(lambda x: x.isna().sum()).reset_index()
```

<strong><h3>Histogramas</h3></strong>

<strong><h3>Correlación</h3></strong>
<p>Las correlaciones se refieren a datos con algún tipo de asociación que evalúa la tendencia creciente o decreciente de los datos. Es muy útil para medir tendencias lineales y saber cómo funciona algo. Para la creación de las correlaciones, utilizamos los métodos de Pearson y Kendall utilizando <strong>.corr(method='pearson') y .corr(method='kendall')</strong>

<strong>Método de Pearson</strong>

```python
lluvias.corr(method = 'pearson')
```

<strong>Método de Kendall</strong>

```python
lluvias.corr(method = 'kendall')
```

<strong><h3>Boxplots</h3></strong>
<p>La librería matplotlib nos da la posibilidad de tener distintos gráficos a partir de todos nuestros datos y ver si hay cierto algoritmo o relación entre distintas variables, por ejemplo:
<p><strong><i>La lluvia depende de tres factores: la presión atmosférica, la temperatura y, especialmente, la humedad atmosférica.</i></strong> <a href="https://es.wikipedia.org/wiki/Lluvia#:~:text=La%20lluvia%20depende%20de%20tres,de%20nieve%20o%20de%20granizo.">Link</a>
<p>Tomando esas variables y analizándolas podemos ver cómo se relacionan de cierta manera entre sí.
<p>Para realizar el graficado se definirá una cadena de variables de interés, en este caso los factores anteriormente mencionados y en un ciclo se creará cada uno de forma automática sin tener que repetir todo nuevamente.
<p>Para graficar sólo se utiliza <strong>sb.boxplot(x="variable", y="other_variable", data=tu_base_de_datos, palette="Set3")</strong>
<i><p>A causa de crasheos en Python, he decidido sólo tomar alrededor de 3000 datos depurados de la ciudad Sydney. Se importará una nueva base de datos</i>

```python
# Carga de datos
lluvias_sydney = pd.read_csv("weatherAUS_sydney.csv")
lluvias_sydney.Date = pd.to_datetime(lluvias_sydney.Date)
```

```python
tablas = ["Pressure3pm", "Temp3pm", "Humidity3pm"]

for variable in tablas:
    
    ax = sb.boxplot(x="Date", y=variable, data=lluvias_sydney, palette="Set3")
    
    plt.title(variable)
    plt.xlabel("Date")
    plt.ylabel(variable)
    plt.show()
```

<strong><h3>Histogramas detallados</h3></strong>
<p>Crearemos unos histogramas para realizar una relación de datos, de esta forma sabremos lo que ocurre cuando existe cierta cantidad de presión, temperatura y humedad para que exista la lluvia.
<p>Se cambiarán los strings yes/no a 0's y 1's de la tabla RainToday para poder hacerlo utilizando .map({'Yes': 1, 'No': 0})
<p>Utilizaremos la funcoón .hist(column=(columna1), by="columna2") para generar los histogramas de las distintas condiciones que hicieron que lloviera.
<p>Se representa la humedad, presión y temperatura con distintos valores vs la cantidad de sol en el día.

```python
lluvias_sydney['RainToday'] = lluvias_sydney['RainToday'].map({'Yes': 1, 'No': 0})
```

```python
for variable in tablas:

    lluvias_sydney.hist(column=(variable), by="RainToday", grid="false", legend=variable)
```

<strong><h3>Heatmap</h3></strong>
    <p>Es un gráfico que tiene un conjunto de valores asociado con colores, sirve para determinar de manera visual qué puntos son más bajos o más altos en base a colores en cierta tabla con datos numéricos. De esta manera es sencillo determinar las situaciones en común que generan una causa. En este caso la lluvia.
    
<p>Aquí podemos ver representadas las condiciones que crean la lluvia donde lo más obscuro es lo más probable y lo menos probable es lo más claro.

```python
plt.figure(figsize=(10,8))
sb.heatmap(lluvias_sydney.corr(),annot=True)
```
