## Histogramas

### Posibilidad de aceptación según resultado de examen

Para realizar los histogramas de forma de gráfica de barras, relacionando los valores de resultado de cada tipo de examen y <br>puntaje académico con su posibilidad de aceptación, agrupamos los valores de cada variable en diez grupos distintos, con <br>rangos específicos. De cada grupo se obtuvo el promedio de valores dentro de dicho rango, el cual se despliega finalmente <br>en la gráfica.

```python
#Para cada una de las variables a analizar
for variable in variables:

    #Se obtienen los valores mínimos y máximos
    mini = min(data[variable])
    maxi = max(data[variable])
    
    #Se obtiene el rango necesario para dividir los datos en diez grupos
    delta = (maxi-mini)/10
    
    grupos = {}
    
    for index,row in data.iterrows():
        #Se obtiene el puntaje de examen o puntaje académico
        value = row[variable]
        
        #Se se inicia el primer grupo a revisar a partir del mínimo y según el rango definido previamente
        rango = mini+delta
        
        while rango<maxi:
            
            #Si el valor se encuentra dentro de ese rango
            if value<rango:
                
                #Se crea el espacio para este grupo si no se ha creado antes
                grupos.setdefault(rango,{"suma":0,"cantidad":0})
                
                #Se suma el valor de esta celda al total
                grupos[rango]["suma"]+=row["Chance of Admit "]
                
                #Se mantiene una cantidad de celdas sumadas para poder sacar el promedio
                grupos[rango]["cantidad"]+=1
                break
            
            #Si no se encuentra el valor dentro de este rango, se mueve al siguiente según el delta obtenido
            rango+=delta
        
    #Se realiza el cálculo del promedio de cada grupo según su suma y su cantidad de datos
    for key in grupos:
        grupos[key]=grupos[key]["suma"]/grupos[key]["cantidad"]
    
    #Se despliega la gráfica
    plt.title(variable)
    plt.xlabel(variable)
    plt.ylabel("Posibilidad de aceptación")
    plt.bar(grupos.keys(),grupos.values(),width=delta,ec="white",align="center")
    plt.show()
    
```

## Mapas de Calor

Por medio del mapa de calor, se pudo obtener una demostración gráfica de la relación entre tres variables distintas. En este <br>caso, se compararon los resultados de cada examen o puntaje académico, en conjunto con el nivel de la escuela a la que se <br>aplica y la probabilidad de aceptación de la combinación de ambos elementos. 

### Posibilidad de aceptación según examen y nivel de la escuela

```python
#Para cada resultado de examen o puntaje académico
for variable in variables:
    
    #Se crea un diccionario donde se relacionará la posibilidad de aceptación con cada combinación de puntaje y
    #calidad de universidad
    dic = {}
    print(variable)
    
    #Para cada una de las columnas de la tabla
    for index,row in data.iterrows():
        
        #Se obtienen las columnas a relacionar, según la variable actualmente analizada
        rating = int(row["University Rating"])
        result = row[variable]
        chance = row["Chance of Admit "]
        
        #Se crea un espacio en la interconección entre el puntaje y la calidad de universidad para obtener el promedio
        #de todos los elementos que se encuentren en esa intersección
        dic.setdefault(rating,{}).setdefault(result,{"suma":0,"cantidad":0})
        
        #Se agrega la posibilidad a la suma total y se aumenta uno a la cantidad de valores sumados
        dic[rating][result]["suma"]+=chance
        dic[rating][result]["cantidad"]+=1
    
    #Para cada lista de valores se obtiene el promedio según su suma y su cantidad de valores
    for key in dic:
        for key_2 in dic[key]:
            dic[key][key_2]=dic[key][key_2]["suma"]/dic[key][key_2]["cantidad"]
    
    #Se ordena el diccionario de menor a mayor según la calidad de la universidad
    ordered = {}
    for i in range(1,6):
        ordered[i] = dic[i]
    
    #Se obtiene la tabla a partir del diccionario
    new = pd.DataFrame.from_dict(ordered)
    
    #Se asigna un cero a todos los espacios vacíos
    new = new.fillna(0)
    
    #Se genera el mapa de calor
    ax= sb.heatmap(new)
    
    #Se muestra una previsualización de los datos
    display(new.head())
    
    #Se publica la gráfica
    plt.title("Posibilidad de aceptación según "+variable)
    plt.xlabel("Nivel de la Escuela")
    plt.ylabel(variable)
    plt.show()
    
```

# Clustering

```python
#Se importan las librerías necesarias para el clustering
import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix, classification_report

#Se le aplica una escala a los datos 
X = scale(data)

plt.figure(figsize=(10,10))

#Los colores a utilizar en los mapas
color_theme = np.array(['darkgray','lightsalmon','powderblue','mediumseagreen'])

#---Agrupación entre TOEFL y GRE, según la posibilidad de aceptación
#Resultado de tres grupos
clustering = KMeans(n_clusters=3,random_state = 0)
clustering.fit(X)
plt.subplot(2,2,1)
plt.scatter(x=data["TOEFL Score"],y=data["GRE Score"],c=color_theme[clustering.labels_],s=30)
plt.title("KMeans TOEFL y GRE (3 Grupos)")

#Resultado de cuatro grupos
clustering = KMeans(n_clusters=4,random_state = 0)
clustering.fit(X)
plt.subplot(2,2,2)
plt.scatter(x=data["TOEFL Score"],y=data["GRE Score"],c=color_theme[clustering.labels_],s=30)
plt.title("KMeans TOEFL y GRE (4 Grupos)")

#---Agrupación entre TOEFL y CGPA, según la posibilidad de aceptación
#---Resultado de tres grupos
clustering = KMeans(n_clusters=3,random_state = 0)
clustering.fit(X)
color_theme = np.array(['darkgray','lightsalmon','powderblue','mediumseagreen','pink'])
plt.subplot(2,2,3)
plt.scatter(x=data["CGPA"],y=data["TOEFL Score"],c=color_theme[clustering.labels_],s=30)
plt.title("TOEFL y CGPA (3 Grupos)")

#Resultado de cuatro grupos
clustering = KMeans(n_clusters=4,random_state = 0)
clustering.fit(X)
plt.subplot(2,2,4)
plt.scatter(x=data["CGPA"],y=data["TOEFL Score"],c=color_theme[clustering.labels_],s=30)
plt.title("KMeans TOEFL y CGPA (4 Grupos)")
plt.show()
```
