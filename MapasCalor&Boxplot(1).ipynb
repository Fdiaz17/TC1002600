```python
fechas = np.array(base["IMDb rating"])[:40]
#extracting the total ratings for each year
overall_rating = []
for fecha in fechas:
    tot_rates = []
    for i in range(len(base["IMDb rating"])):
        if base["IMDb rating"][i] == fecha:
            tot_rates.append(base['No of seasons available'][i])
    overall_rating.append(tot_rates[:30])
data = pd.DataFrame(overall_rating).T
data.columns = fechas
#Plotting
fig = plt.figure()
ax = fig.subplots()
x=sb.heatmap(data, cmap = 'viridis')
ax.invert_yaxis()
ax.set_xlabel('Calificacion')
ax.set_ylabel(' ')
plt.show()
```
