<h1>Boxplot<h1>

```python
sb.set_theme(style="whitegrid")
iris = sb.load_dataset("iris")
ax = sb.boxplot(x=iris["sepal_length"])
```

```python
ax = sb.boxplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
ax = sb.boxplot (x="sepal_length", y="sepal_width", data=iris)
ax = sb.stripplot (x="sepal_length", y="sepal_width", data=iris, color="0.0000003")
```

```python
g = sb.catplot(x="sepal_length", y="petal_length",
            hue="species", col="species",
            data=iris, kind="box",
            height=4, aspect=.7);
```
