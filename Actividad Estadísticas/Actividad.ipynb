## Duración
Al observar las gráficas originales obtenidas por los datos de duración, observamos que existían datos atípicos, que mostraban <br> duraciones demasiado extensas que, al verificar en internet, demostraban ser incorrectas. Por lo <br> tanto, para corregirlo, realizamos un nuevo arreglo, con datos que estuvieran en un rango específico para así eliminar <br> todos los datos que afectaran demasiado la gráfica final.

<br>Tras realizar esta limpieza, se puede observar una gran diferencia en los datos, ya que la media se encontraba cien <br> segundos más arriba de lo en realidad observado en la gráfica, y la desviación estandar era de hasta seiscientos segundos, en <br> lugar de los tan solo sesenta que se observan en los datos tras la limpieza. Realizar este filtro de datos nos permitió <br> describir los datos de una manera más apegada a lo requerido.
<br>
<br> Al ser la única variable con datos numéricos, fue aquella la única a la que pudimos realizarle las descripciones de <br> mínimo, máximo, media y desviación estandar.
