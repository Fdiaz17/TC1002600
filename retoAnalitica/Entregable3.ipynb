# Diagramas de caja y bigote
Primero compararemos la cantidad de likes obtenidas en cada categoría, apoyándonos de diagramas de cajas

```python

sb.set_theme(style = "whitegrid")
sbdata = cutdown2

ax = sb.boxplot (x="category_id", y= "likes", data = sbdata)
plt.title('Cantidad de likes por categoría')
```

```python
ax = sb.boxplot (x="category_id", y= "likes", data = sbdata)

ax.set(ylim=(0, 70000))
plt.title('Cantidad de likes por categoría')
```

```python
ax = sb.boxplot (x="category_id", y= "dislikes", data = sbdata)
plt.title('Cantidad de dislikes por categoría')
```

Realizamos un acercamiento a la gráfica para apreciar mejor los datos

```python
ax = sb.boxplot (x="category_id", y= "dislikes", data = sbdata)

ax.set(ylim=(0, 2000))
plt.title('Cantidad de dislikes por categoría')
```

Utilizando la base de datos yearmonth, la cual contiene las estadísticas de un video aleatorio por mes,compararemos estos datos

Cabe mencionar que para que los datos sean más significativos sería necesario calcular el promedio por año, sin embargo debido al tamaño de la base de datos original (+40,000 registros) la complejidad de catalogar manualmente cada video por mes y año para después sacar promedio queda fuera del alcance.

```python
ax = sb.boxplot(x="year", y= "likes", data = yearmonth)

ax.set(ylim=(0, 10000))
plt.title('Cantidad de likes en el video de cada año')
```

# Histogramas
En este histograma se compara la cantidad de videos que reciben n cantidad de likes

# Mapas de calor

Elegimos utilizar un mapa de calor con la intención de representar la cantidad de visualizaciones que tuvo algún video correspondiente a un mes y año 
dentro de la base de datos.
Para esto, filtramos preivamente la información y la guardamos en year-month-stats.csv.
Cabe mencionar que la base de datos no cuenta con videos de todos los meses para cada año, a excepción de 2017

```python
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
yticks_labels = ['2009', '2010', '2011', '2012', '2013',
                 '2014','2015', '2016', '2017']

ax = sb.heatmap(heatprep, vmax=100000, square = True, yticklabels=yticks_labels, xticklabels= xticks_labels)
plt.title('Cantidad de views de un video por mes en cada año')
plt.ylabel('year')
plt.xlabel('month')
ax.xaxis.tick_top()

plt.show()
```
